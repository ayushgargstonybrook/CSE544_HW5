```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
x=np.genfromtxt('A5_q2.dat',dtype="double",delimiter=",")
xlist=[]
ylist=[]
for i in range(x.shape[0]):
    xlist.append(x[i][0])
    ylist.append(x[i][1])
```

```python
sse=0
mape=0
ypred=[]
for i in range(len(xlist)):
    ypr=x[i][0]
    ypred.append(ypr)
    mape=mape+abs(ypr-x[i][1])/x[i][1]
    sse=sse+(ypr-x[i][1])*(ypr-x[i][1])
mape=mape/len(ypred)
print("mape is ",mape," %")
print("sse is ",sse)
plt.plot(xlist,ylist,label="original")
plt.plot(xlist,ypred,label="regression( Y = X )")
plt.xlabel('server CPU usage percentage(X)')
plt.ylabel('response time(Y)')
pylab.legend(loc='upper left')   
plt.show() 
```
